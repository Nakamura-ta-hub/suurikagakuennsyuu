```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure()
x = np.arange(0, 1, 0.01)
ims = []
for i in range(50):
    t=0.01*i
    u = 0.5*np.exp(-100*(x+t-0.5)**2)+0.5*np.exp(-100*(x-t-0.5)**2)
    im = plt.plot(x, u, "r")
    ims.append(im)
    
    ani = animation.ArtistAnimation(fig, ims)
plt.xlabel("x") #　横軸のラベル
plt.ylabel("u") #　縦軸のラベル

plt.show()
```

コメント

0.5を中心として、が増大するにつれ、正負逆方向に拡散していく。
uの積分はどのt地点においても一定。
モデル化ってすごい。
